```python
import pandas as pd
from textblob import TextBlob
!pip install googletrans==4.0.0-rc1
from googletrans import Translator

```

```python
translator = Translator()

def translateComment(strComment: str):
    strEngComment = translator.translate(strComment, src="es", dest="en").text
    return strEngComment

def getPolarity(strComment: str):
    strEngComment = translateComment(strComment)
    analysis = TextBlob(strEngComment)
    return analysis.sentiment.polarity

# Leer el archivo CSV
df = pd.read_csv('/content/Comentarios.csv')

# Aplicar la función de polaridad
df['polaridad'] = df['Comentario'].apply(getPolarity)

# Guardar los resultados en un nuevo archivo CSV
df.to_csv('resultados.csv', index=False)
```
